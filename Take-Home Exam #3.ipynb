```python
import re
import nltk 
from nltk.corpus import *
corpus_root = "C:/Users/sana/인공지능/raw_text2/"
GNI = PlaintextCorpusReader(corpus_root,".*\.txt", encoding = "utf-8")
giRaw = GNI.raw()  
```

# 대문자만으로 된 약어 

```python
list1 = re.findall(r"\s[A-Z]{2,}\s",giRaw)
print ("전체 대문자 약어 수 : ",len(list1))
print (list1)
```

```python
list1 = list(set(list1))
print ("전체 대문자 약어 수 (중복제거) : ",len(list1))
print (list1)
```

# 알파벳으로 시작하고 숫자로 끝나는 특수 용어

```python
list2 = re.findall(r"\s[a-zA-Z]\w*\d\s",giRaw)
print ("전체 알파벳으로 시작하고 숫자로 끝나는 특수 용어 수 : ",len(list2))
print(list2)
```

```python
list2 = list(set(list2))
print ("전체 대문자 약어 수 (중복제거) : ",len(list2))
print (list2)
```

# 그 밖의 특수 용어

```python
list3 = re.findall(r"\s[(]\w*[)]\s",giRaw)
print ("그 밖의 특수 용어 - () 로 둘러싸인 대소문자 숫자로만 이루어진 용어 수 : ",len(list3))
print(list3)
```

```python
list3 = list(set(list3))
print ("그 밖의 특수 용어 - () 로 둘러싸인 대소문자 숫자로만 이루어진 용어 수 (중복제거) : ",len(list3))
print (list3)
```
